```python
# Cleaning up Final Matches Scores
# adding penalty column. 4-3 pen. 0-0 (0-0, 0-0)

#copy to penalty
#f2["penalty"] = f2["score"]

def get_score(item):
    if 'pen' in item:
        pen = item.split()[0].split("-")
        p = True
        pen_h = pen[0]
        pen_a = pen[1]
        sc = item.split()[2].split("-")
    else:
        p = False
        pen_h = 0
        pen_a = 0
        sc = item.split()[0].split("-")
    sc_h = sc[0]
    sc_a = sc[1]
    return (sc_h, sc_a, pen_h, pen_a, p)
```

```python
#getting score and mapping it to respective coloumns
p = f2["score"].map(get_score)

f2["home_score"]= f2["score"].map(lambda x: get_score(x)[0])

f2["away_score"]= f2["score"].map(lambda x: get_score(x)[1])

f2["home_penalty"]= f2["score"].map(lambda x: get_score(x)[2])

f2["away_penalty"]= f2["score"].map(lambda x: get_score(x)[3])

f2["penalty"] = f2["score"].map(lambda x:get_score(x)[4])

```

```python
#Concat f3 and f2 to create 1 table of entries

f3["home_penalty"] = 0
f3["away_penalty"] = 0

f3["penalty"] = False

f4 = pd.concat([f3, f2])
f4["home_score"] = f4["home_score"].astype('int')
f4["away_score"] = f4["away_score"].astype('int')
f4["home_penalty"] = f4["home_penalty"].astype('int')
f4["away_penalty"] = f4["away_penalty"].astype('int')
```

```python
def parse_player(entry, match_num, team):
    lm = []
    tokens = entry.split()
    i = 0
    ln = len(tokens)
    while i < ln:
        player = ""
        #Collect name
        while(not (tokens[i][0] in st.digits)):
            player = player + " " + tokens[i]
            i = i + 1
        player = player.strip()
        while(i < ln and tokens[i][0] in st.digits):
            tks = tokens[i].split("'")[0].split("+")
            time = tks[0]
            if len(tks) > 1:
                stopage = tks[1]
            else:
                stopage = 0
            i = i + 1
            penalty = False
            own = False
            if i < ln and "pen" in tokens[i]:
                penalty = True
                i = i + 1
            elif i < ln and "o.g." in tokens[i]:
                own = True
                i = i + 1
            lm.append((match_num, player, time, penalty, own, stopage, team))
    return lm

for rowt in f4.iterrows():
    row = rowt[1]
    if row["home_score"] > 0:
        try:
            lm = parse_player(row["home_scorers"], row["match_num"], "Home")
            lt = lt + lm
        except:
            print(row["home_scorers"])
            print("err")
    if row["away_score"] > 0: 
        try:
            lm = parse_player(row["away_scorers"], row["match_num"], "Away")
            lt = lt + lm
        except:
            print(row["away_scorers"])
            print("err")

Goals = pd.DataFrame(lt, columns=["match_num", "Player", "Time", "Penalty",
    "Own_goal",
    "Stopage_Time", "Home/Away"])

Matches = f4
Dates = f1

#TABLES Required
print(Goals.head())
print(Matches.head())
print(Dates.head())
print(team_to_grp.head())
```

```python
points_map = {
    'W': 3,
    'D': 1,
    'L': 0
}

def get_result(score, score_opp):
    if score == score_opp:
        return 'D'
    elif score > score_opp:
        return 'W'
    else:
        return 'L'
```

```python
def get_dominating(score, score_opp):
        dominate= (score-score_opp)
        return dominate
```

```python
# full time goals
team_grp_rslt_df['Goals'] = np.where(team_grp_rslt_df['Team'] == team_grp_rslt_df['home'],
                                 team_grp_rslt_df['home_score'],
                                 team_grp_rslt_df['away_score'])
team_grp_rslt_df['Goals_Opp'] = np.where(team_grp_rslt_df['Team'] != team_grp_rslt_df['home'],
                                     team_grp_rslt_df['home_score'],
                                     team_grp_rslt_df['away_score'])
team_grp_rslt_df['Result'] = np.vectorize(get_result)(team_grp_rslt_df['home_score'], team_grp_rslt_df['away_score'])
team_grp_rslt_df['Points'] = team_grp_rslt_df['Result'].map(points_map)
```
